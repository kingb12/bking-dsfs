```python
from matplotlib import pyplot as plt
from collections import Counter
import random
import seaborn
import pandas
```

```python
# random gdp data over 10 decades
random.seed(42)
years = [1950 + 10*i for i in range(0, 10)]
gdp = [i*35 + 300 + random.randrange(-500, 500) + random.random() for i in range(0, 10)]
```

```python
plt.plot(years, gdp, color='green', marker='o', linestyle='solid')
plt.title('Random GDP data')
plt.ylabel('Billions of $')
plt.show()
```

```python
# random oscar data for a bar chart
movies = ['The Dark Knight', '2001: Space Oddyssey', 'The Big Lebowski', '2012', 'Fargo']
num_oscars = [2, 2, 200, 0, 100]
```

```python
# bar chart configuration
plt.bar(range(len(movies)), num_oscars)
plt.title('Some Neat Movies')
plt.ylabel('# of times I told Reilly about them')
plt.xticks(range(len(movies)), movies) # why do I have to do this in such an ugly way?
plt.show()
```

```python
grades = [random.randrange(0, 100) for i in range(0, 30)]
#bucketing grades by decile, putting 100 in with the 90s
histogram = Counter(min(grade//10 * 10, 90) for grade in grades)
# I expect histogram to contain be a dict {0: x_1, 10: x_2 ... 90: x_10}
plt.bar([ x + 5 for x in histogram.keys()], histogram.values(), 10, edgecolor='black')
plt.axis([-5, 105, 0, 10])
# similar to keys but we want to add space for 0, 100 AND we can't guarantee all deciles used
plt.xticks([10 * i for i in range(11)])
plt.xlabel('Decile')
plt.ylabel('Number of Students')
plt.title('Distribution of Exam 1 Grades')
plt.show()
```

```python
# using seaborn for the following scatterplot
visits = [random.randrange(0, 250) for i in range(100)]
voting_for_bernie = [max(0, min(1, (1.0 - float(i) / 250) + random.gauss(0, .10))) for i in visits]
voting_for_biden = [max(0, min(1, (float(i) / 250) + random.gauss(0, .10))) for i in visits]
# need to turn this into a pandas dataframe
data = {'visits': visits, 'bernie': voting_for_bernie, 'biden': voting_for_biden}
df = pandas.DataFrame(data=data)
seaborn.scatterplot(x='visits', y='bernie', data=data)
plt.title('Bernie Vote Likelihood vs. CNN views')
plt.xlabel('CNN Views / month')
plt.ylabel('Likelihood')
plt.axis([-5, 255, 0, 1])
```
